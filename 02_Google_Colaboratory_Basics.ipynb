## 01. Variables

Rules for Python variables:  

1. A variable name must start with a letter or the underscore character  
2. A variable name cannot start with a number  
3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )  
4. Variable names are case-sensitive (age, Age and AGE are three different variables)  

```
# Creating Variables
x = 1
y = 'GitHub'
z = 8.88
print(x, y, z)
print(type(x), type(y), type(z))
```

```
# Casting / Changing variable type
x = float(10)
y = int("20")
z = int(30.123)
a = str(40)
print(x, y, z, a)
print(type(x), type(y), type(z), type(a))
```

```
# String operations
# indexing starts with 0

a = "Hello, World!"
print(a[0])

b = "Hello, World!"
print(b[2:5])  

c = "   Hello, World!   " 
print(c.strip())

a = "Hello, World!"
print('Length:', len(a))

a = "Hello, World!"
print(a.lower())

a = "Hello, World!"
print(a.upper())

a = "Hello, World!"
print(a.replace("H", "A"))

a = "Hello, World!"
print(a.split(",")) 
```

## 04.Lists and Dictionaries

```
thislist = ["apple", "banana", "cherry"]  # This is a list
print(len(thislist))  # Length of the list
```

```
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")   # Add an item to the end of the list
print(thislist)
```

```
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")   # Add an item to a specific location in the list
print(thislist)
```

```
thislist = ["apple", "banana", "cherry"]
del thislist[0]   # Delete a selected item from the list
print(thislist)
```

```
# Sort a list
cars = ['Ford', 'BMW', 'Volvo']
cars = sorted(cars, reverse=False)
print(cars)
```

```
# Loop through a list
fruit = ["apple", "banana", "cherry"]
for x in fruit:
  print(x)
```

```
fruit = ["apple", "banana", "cherry"]

# Check if an element is in the list
if "apple" in fruit:
  print("Yes, 'apple' is in the fruit list")
```

### 06.01Lambda functions

A lambda function is a straightforward anonymous function. A lambda function can take any number of arguments, but can only have one expression.

```
# Take this function
def f(a):
  return a + 10
print(f(19))
```

```
# Above function is similar to following;
f = lambda a : a + 10
print(f(19))
```

```
# Following lambda function is similar to;
# def z(a, b):
#   return a+b

z = lambda a, b : a + b
print(z(3, 8))
```
